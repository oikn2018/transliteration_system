```python
# Reference : https://keras.io/examples/nlp/lstm_seq2seq/
input_characters = set()
target_characters = set()
input_characters.add(' ')
target_characters.add(' ')
```

```python
#For train data
input_words_train = []
target_words_train = []

for j in range(len(train_data)):
  try:
    input_word_train=train_data.iloc[j,1]
    target_word_train=train_data.iloc[j,0]
    target_word_train = "\t" + target_word_train + "\n"
    if(len(input_word_train)>=1 and len(target_word_train)>=2):
      input_words_train.append(input_word_train)
      target_words_train.append(target_word_train)
    for char in input_word_train:
      if char not in input_characters:
        input_characters.add(char)
    for char in target_word_train:
      if char not in target_characters:
        target_characters.add(char)
  except: continue
```

```python
#For Validation data:
input_words_val = []
target_words_val = []

for j in range(len(validation_data)):
  try:
    input_word_val=validation_data.iloc[j,1]
    target_word_val=train_data.iloc[j,0]
    target_word_val = "\t" + target_word_val + "\n"
    if(len(input_word_val)>=1 and len(target_word_val)>=2):
      input_words_val.append(input_word_val)
      target_words_val.append(target_word_val)
    for char in input_word_val:
      if char not in input_characters:
        input_characters.add(char)
    for char in target_word_val:
      if char not in target_characters:
        target_characters.add(char)
  except: continue
```

```python
#For test data:
input_words_test = []
target_words_test = []

for j in range(len(test_data)):
  try:
    input_word_test=test_data.iloc[j,1]
    target_word_test=test_data.iloc[j,0]
    target_word_test = "\t" + target_word_test + "\n"
    if(len(input_word_test)>=1 and len(target_word_test)>=2):
      input_words_test.append(input_word_test)
      target_words_test.append(target_word_test)
    for char in input_word_test:
      if char not in input_characters:
        input_characters.add(char)
    for char in target_word_test:
      if char not in target_characters:
        target_characters.add(char)
  except: continue
```

```python
input_characters       = sorted(list(input_characters))
target_characters      = sorted(list(target_characters))
num_encoder_tokens = len(input_characters)
num_decoder_tokens = len(target_characters)
```

```python
max_encoder_seq_length = max(max([len(text) for text in input_words_train]), max([len(text) for text in input_words_val]), max([len(text) for text in input_words_test]))
max_decoder_seq_length = max(max([len(text) for text in target_words_train]), max([len(text) for text in target_words_val]), max([len(text) for text in target_words_test]))
```

```python
print("Number of train samples: ", len(input_words_train))
print("Number of validation samples: ", len(input_words_val))
print("Number of test samples: ", len(input_words_test))
print("Number of unique input tokens: ", num_encoder_tokens)
print("Number of unique output tokens: ", num_decoder_tokens)
print("Max sequence length for inputs: ", max_encoder_seq_length)
print("Max sequence length for train outputs: ", max_decoder_seq_length)
```

```python
input_token_index = dict([(char, i) for i, char in enumerate(input_characters)])
target_token_index = dict([(char, i) for i, char in enumerate(target_characters)])

reverse_input_char_index = dict((i, char) for char, i in input_token_index.items())
reverse_target_char_index = dict((i, char) for char, i in input_token_index.items())
```

```python
encoder_train_input_data = np.zeros(
    (len(input_words_train), max_encoder_seq_length), dtype="float32"
    )
decoder_train_input_data = np.zeros(
    (len(input_words_train), max_decoder_seq_length), dtype="float32"
    )
decoder_train_target_data = np.zeros(
    (len(input_words_train), max_decoder_seq_length, num_decoder_tokens ), dtype="float32"
    )

encoder_val_input_data = np.zeros(
    (len(input_words_val), max_encoder_seq_length), dtype="float32"
    )
decoder_val_input_data = np.zeros(
    (len(input_words_val), max_decoder_seq_length), dtype="float32"
    )
decoder_val_target_data = np.zeros(
    (len(input_words_val), max_decoder_seq_length, num_decoder_tokens), dtype="float32"
    )

encoder_test_input_data = np.zeros(
    (len(input_words_test), max_encoder_seq_length), dtype="float32"
    )
```

```python
for i, (input_word, target_word) in enumerate(zip(input_words_train, target_words_train)):
    for t, char in enumerate(input_word):
        encoder_train_input_data[i, t] = input_token_index[char]
    encoder_train_input_data[i, t + 1 :] = input_token_index[' ']
        
    for t, char in enumerate(target_word):
        #decoder_train_target_data is ahead of decoder_train_input_data by one timestamp
        decoder_train_input_data[i, t] = target_token_index[char]
        if t > 0:
        # decoder_train_target_data will be ahead by one timestep
        # and will not include the start character.

            decoder_train_target_data[i, t - 1, target_token_index[char]] = 1.0
    decoder_train_input_data[i, t + 1 :] = target_token_index[' ']
    decoder_train_target_data[i, t :, target_token_index[' ']] = 1.0
```

```python
for i, (input_word, target_word) in enumerate(zip(input_words_val, target_words_val)):
    for t, char in enumerate(input_word):
        encoder_val_input_data[i, t] = input_token_index[char]
    encoder_val_input_data[i, t + 1 :] = input_token_index[' ']
        
    for t, char in enumerate(target_word):
        #decoder_val_target_data is ahead of decoder_val_input_data by one timestamp
        decoder_val_input_data[i, t] = target_token_index[char]
        if t > 0:
        # decoder_val_target_data will be ahead by one timestep
        # and will not include the start character.

            decoder_val_target_data[i, t - 1, target_token_index[char]] = 1.0
    decoder_val_input_data[i, t + 1 :] = target_token_index[' ']
    decoder_val_target_data[i, t :, target_token_index[' ']] = 1.0
```

```python
for i, input_word in enumerate(input_words_test):
    for t, char in enumerate(input_word):
        encoder_test_input_data[i, t] = input_token_index[char]
    encoder_test_input_data[i, t + 1 :] = input_token_index[' ']
#print(type(encoder_test_input_data))    
```

```python
# Configure the sweep
sweep_config = {
    'method': 'bayes', 
    'metric': {
      'name': 'accuracy',
      'goal': 'maximize'   
    },
    'parameters': {
        "encoder_layers" :{
            "values" : [1, 2, 3]
        },
        "decoder_layers": {
            "values": [1, 2, 3]
        },
        "hidden_layer_size": {
            "values": [16, 32, 64, 256]
        },
        "dropout": {
            "values": [0.2, 0.3]
        },

        'cell_type':{
            'values':['RNN','GRU','LSTM']
        },
        'train_word_accuracy':{
             'values':[0]
        },
        'input_embedding_size':{
             'values':[16,32,64,256]
        } 
        

        
    }
}
```

```python
def train(config = None):
    # Initialize a new wandb run
    run = wandb.init(config = config)
    config = wandb.config


    run.name =  " Cell Type: " + config.cell_type + "Embedding Size: " + str(config.input_embedding_size) +" Dropout: " + str(config.dropout) + " train_word_accuracy: " + str(config.train_word_accuracy) + " Encoder Layers: " + str(config.encoder_layers) + " Decoder Layers: " + str(config.decoder_layers) + " Hidder Layer Size: " + str(config.hidden_layer_size)

    # Configuration defaults
    epochs = 20
    batch_size = 64   
    latent_dim_encoder = [config.hidden_layer_size] * config.encoder_layers
    latent_dim_decoder  = [config.hidden_layer_size] * config.decoder_layers
    cell_type = config.cell_type
    dropout = config.dropout
    train_word_accuracy = config.train_word_accuracy
    input_embedding_size = config.input_embedding_size


    # Encoder for the model
    encoder_inputs = keras.Input(shape = (None, ))
    encoder_outputs = keras.layers.Embedding(input_dim = num_encoder_tokens, output_dim = input_embedding_size, input_length = max_encoder_seq_length)(encoder_inputs)

    
    encoder_states = list()
    #different cells such as vanilla RNN, LSTM and GRU 
    for j in range(len(latent_dim_encoder)):
        if cell_type == "RNN":
            encoder_outputs, state = keras.layers.SimpleRNN(latent_dim_encoder[j], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state]
        if cell_type == "LSTM":
            encoder_outputs, state_h, state_c = keras.layers.LSTM(latent_dim_encoder[j], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state_h,state_c]
        if cell_type == "GRU":
            encoder_outputs, state = keras.layers.GRU(latent_dim_encoder[j], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state]

    # Decoder
    decoder_inputs = keras.Input(shape=(None, ))
    decoder_outputs = keras.layers.Embedding(input_dim = num_decoder_tokens, output_dim = input_embedding_size, input_length = max_decoder_seq_length)(decoder_inputs)


    decoder_states = encoder_states.copy()

    for j in range(len(latent_dim_decoder)):
        if cell_type == "RNN":
            decoder = keras.layers.SimpleRNN(latent_dim_decoder[j], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state = decoder(decoder_outputs, initial_state = decoder_states)
      
        if cell_type == "LSTM":
            decoder = keras.layers.LSTM(latent_dim_decoder[j], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state_h, state_c = decoder(decoder_outputs, initial_state = decoder_states)
           
        if cell_type == "GRU":
            decoder = keras.layers.GRU(latent_dim_decoder[j], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state = decoder(decoder_outputs, initial_state = decoder_states)
           

    decoder_dense = keras.layers.Dense(num_decoder_tokens, activation = "softmax")
    decoder_outputs = decoder_dense(decoder_outputs)

 
    model = keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)

    model.compile(
        optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"]
    )

    model.fit(
        [encoder_train_input_data, decoder_train_input_data],
        decoder_train_target_data,
        batch_size = batch_size,
        epochs = epochs,
        callbacks = [WandbCallback()]
    )

    model.save("vanilla_seq2seq_model")
   
```
