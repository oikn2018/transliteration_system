```python
# Reference : https://keras.io/examples/nlp/lstm_seq2seq/
input_characters = set()
target_characters = set()

input_characters.add(' ')
target_characters.add(' ')
```

```python
#For Train data
input_texts_train = []
target_texts_train = []
for text in list(train_data[1]):
    input_texts_train.append(str(text))
for text in list(train_data[0]):
     target_texts_train.append("\t" + str(text) + "\n")   

print(target_texts_train)
print(len(target_texts_train))

for train_text in input_texts_train:
    for char in train_text:
        input_characters.add(char)

for train_text in target_texts_train:
    for char in train_text:
        target_characters.add(char)      
```

```python
#For Validation data
input_texts_val = []
target_texts_val = []
for text in list(validation_data[1]):
    input_texts_val.append(str(text))
for text in list(validation_data[0]):
     target_texts_val.append("\t" + str(text) + "\n")   

print(target_texts_val)
print(len(target_texts_val))

for val_text in input_texts_val:
    for char in val_text:
        input_characters.add(char)

for val_text in target_texts_val:
    for char in val_text:
        target_characters.add(char) 
```

```python
#For Test data
input_texts_test = []
target_texts_test = []
for text in list(test_data[1]):
    input_texts_test.append(str(text))
for text in list(test_data[0]):
     target_texts_test.append("\t" + str(text) + "\n")   

print(target_texts_test)
print(len(target_texts_test))

for test_text in input_texts_test:
    for char in test_text:
        input_characters.add(char)

for test_text in target_texts_test:
    for char in test_text:
        target_characters.add(char) 
```

```python
input_characters  = sorted(list(input_characters))
target_characters = sorted(list(target_characters))
num_encoder_tokens = len(input_characters)
num_decoder_tokens = len(target_characters)

max_encoder_seq_length = max(max([len(word) for word in input_texts_train]), max([len(word) for word in input_texts_val]), max([len(word) for word in input_texts_val]))
max_decoder_seq_length = max(max([len(word) for word in target_texts_train]), max([len(word) for word in target_texts_val]), max([len(word) for word in target_texts_test]))
```

```python
# Printing the summary
print("Number of train samples: ", len(input_texts_train))
print("Number of validation samples: ", len(input_texts_val))
print("Number of test samples: ", len(input_texts_test))
print("Number of unique input tokens: ", num_encoder_tokens)
print("Number of unique output tokens: ", num_decoder_tokens)
print("Max sequence length for inputs: ", max_encoder_seq_length)
print("Max sequence length for train outputs: ", max_decoder_seq_length)
```

```python
input_token_index = dict([(char, i) for i, char in enumerate(input_characters)])
target_token_index = dict([(char, i) for i, char in enumerate(target_characters)])

reverse_input_char_index = dict((i, char) for char, i in input_token_index.items())
reverse_target_char_index = dict((i, char) for char, i in input_token_index.items())
```

```python
# data holders
encoder_train_input_data = np.zeros(
        (len(input_texts_train), max_encoder_seq_length), dtype="float32"
    )
decoder_train_input_data = np.zeros(
        (len(input_texts_train), max_decoder_seq_length), dtype="float32"
    )
decoder_train_target_data = np.zeros(
        (len(input_texts_train), max_decoder_seq_length, num_decoder_tokens), dtype="float32"
    )

encoder_val_input_data = np.zeros(
        (len(input_texts_val), max_encoder_seq_length), dtype="float32"
    )
decoder_val_input_data = np.zeros(
        (len(input_texts_val), max_decoder_seq_length), dtype="float32"
    )
decoder_val_target_data = np.zeros(
        (len(input_texts_val), max_decoder_seq_length, num_decoder_tokens), dtype="float32"
    )

encoder_test_input_data = np.zeros(
        (len(input_texts_test), max_encoder_seq_length), dtype="float32"
    )
```

```python
# Train data
for i, (input_text, target_text) in enumerate(zip(input_texts_train, target_texts_train)):
    for t, char in enumerate(input_text):
        encoder_train_input_data[i, t] = input_token_index[char]
    encoder_train_input_data[i, t + 1 :] = input_token_index[' ']
        
    for t, char in enumerate(target_text):

        decoder_train_input_data[i, t] = target_token_index[char]
        if t > 0:
        # decoder's train target data will be ahead by one timestep
        # And will not include the start character.

            decoder_train_target_data[i, t - 1, target_token_index[char]] = 1.0
    decoder_train_input_data[i, t + 1 :] = target_token_index[' ']
    decoder_train_target_data[i, t :, target_token_index[' ']] = 1.0
```

```python
# Validation data
for i, (input_text, target_text) in enumerate(zip(input_texts_val, target_texts_val)):
    for t, char in enumerate(input_text):
        encoder_val_input_data[i, t] = input_token_index[char]
    encoder_val_input_data[i, t + 1 :] = input_token_index[' ']
        
    for t, char in enumerate(target_text):
        decoder_val_input_data[i, t] = target_token_index[char]
        if t > 0:
        # decoder's validation target data will be ahead by one timestep
        # And will not include the start character.

            decoder_val_target_data[i, t - 1, target_token_index[char]] = 1.0
    decoder_val_input_data[i, t + 1 :] = target_token_index[' ']
    decoder_val_target_data[i, t :, target_token_index[' ']] = 1.0
```

```python
for i, input_text in enumerate(input_texts_test):
    for t, char in enumerate(input_text):
        encoder_test_input_data[i, t] = input_token_index[char]
    encoder_test_input_data[i, t + 1 :] = input_token_index[' ']
#print(type(encoder_test_input_data))   
```

Sweep Configuration 

```python

# Configure the sweep
sweep_config = {
    'method': 'bayes', 
    'metric': {
      'name': 'accuracy',
      'goal': 'maximize'   
    },
    'parameters': {
        "encoder_layers" :{
            "values" : [1, 2, 3]
        },
        "decoder_layers": {
            "values": [1, 2, 3]
        },
        "hidden_layer_size": {
            "values": [16, 32, 64, 256]
        },
        "dropout": {
            "values": [0.2, 0.3]
        },
        "beam_size": {
            "values": [0,2]
        },

        'cell_type':{
            'values':['rnn','gru','lstm']
        },

        'input_embedding_size':{
             'values':[16,32,64,256]
        } 
        

        
    }
}
```

```python
# Train Function
def train(config = None):
    run = wandb.init(config = config)
    config = wandb.config

    # Run Name
    run.name = " Cell: " + config.cell_type + "Input_Embedding: " + str(config.input_embedding_size)  + " Beam_size: " + str(config.beam_size) +" Dropout: " + str(config.dropout) + " Encoder_layers: " + str(config.encoder_layers) + " Decoder_layers: " + str(config.decoder_layers) + " Hidden_layer: " + str(config.hidden_layer_size)

    # Fixed Configuration Parameters:
    epochs = 25
    #batch_size = 64
    batch_size = 128
    
    ''' 
    Variable Configuiration Parameters:
    cell_type -- (string) type of cell used in encoder and decoder (possible values : 'lstm', 'gru', 'rnn')
    model -- (Keras model object) non-attention model used for training
    encoder_model -- (Keras model object) 
    decoder_model -- (Keras model object)
    '''
    cell_type = config.cell_type
    embedding_size = config.input_embedding_size
    beam_size = config.beam_size
    dropout = config.dropout
    latent_dim_encoder = [config.hidden_layer_size] * config.encoder_layers
    latent_dim_decoder  = [config.hidden_layer_size] * config.decoder_layers

    # Initialize place holder for encoder states
    encoder_states = list()
    decoder_states = encoder_states.copy()
    # Define an input and output sequence for encoder and process it.
    encoder_inputs = keras.Input(shape = (None, ))
    encoder_outputs = keras.layers.Embedding(input_dim = num_encoder_tokens, output_dim = embedding_size, input_length = max_encoder_seq_length)(encoder_inputs)
    
    #different cells such as vanilla RNN, LSTM and GRU 
    for i in range(len(latent_dim_encoder)):
        if cell_type == "rnn":
            encoder_outputs, state = keras.layers.SimpleRNN(latent_dim_encoder[i], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state]
        if cell_type == "lstm":
            encoder_outputs, state_h, state_c = keras.layers.LSTM(latent_dim_encoder[i], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state_h,state_c]
        if cell_type == "gru":
            encoder_outputs, state = keras.layers.GRU(latent_dim_encoder[i], dropout = dropout, return_state = True, return_sequences = True)(encoder_outputs)
            encoder_states = [state]

    # Initialize place holder for decoder states
    decoder_states = encoder_states.copy()        
    
    # Define an input and output sequence for decoder and process it.
    decoder_inputs = keras.Input(shape=(None, ))
    decoder_outputs = keras.layers.Embedding(input_dim = num_decoder_tokens, output_dim = embedding_size, input_length = max_decoder_seq_length)(decoder_inputs)
    
    for i in range(len(latent_dim_decoder)):
        if cell_type == "rnn":
            decoder = keras.layers.SimpleRNN(latent_dim_decoder[i], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state = decoder(decoder_outputs, initial_state = decoder_states)
   
        if cell_type == "lstm":
            decoder = keras.layers.LSTM(latent_dim_decoder[i], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state_h, state_c = decoder(decoder_outputs, initial_state = decoder_states)
 
        if cell_type == "gru":
            decoder = keras.layers.GRU(latent_dim_decoder[i], dropout = dropout, return_sequences = True, return_state = True)
            decoder_outputs, state = decoder(decoder_outputs, initial_state = decoder_states)


    decoder_dense = keras.layers.Dense(num_decoder_tokens, activation = "softmax")
    decoder_outputs = decoder_dense(decoder_outputs)    

        
    # Define and build the model    
    model = keras.Model([encoder_inputs, decoder_inputs], decoder_outputs)

    model.compile(
        optimizer="rmsprop", loss="categorical_crossentropy", metrics=["accuracy"]
    )

    model.fit(
        [encoder_train_input_data, decoder_train_input_data],
        decoder_train_target_data,
        epochs = epochs,
        batch_size = batch_size,
        callbacks = [WandbCallback()]
    )

    
    # Inference without attention implementation 

    # It will give the validation accuracy 
    count = 0
    

    # Question it is asked to write predictions_vanilla.csv
    predictions_vanilla = open("predictions_vanilla.csv", "w", encoding='utf-8')
    predictions_vanilla.write("I/P,Predicted,Original\n") 

    
    encoder_inputs = model.input[0]
    decoder_inputs = model.input[1] 
    decoder_states = []
    decoder_states_inputs =  []

    encoder_cell_pos = []
    decoder_cell_pos = []
    encoder_embedding_pos, decoder_embedding_pos,  dense_pos = -1, -1, -1
  

    encoder_cell_index = 0

    #Calculating number of layers in encoder and decoder
    for i, layer in enumerate(model.layers):
    # Search for different layer names (embedding, dense, cell) and increase the pos accordingly    
        if "embedding" in layer.name:
            if encoder_embedding_pos == -1 : 
                encoder_embedding_pos = i
            else: 
                decoder_embedding_pos = i

        if "dense" in layer.name : 
            dense_pos = i

        # rnn, gru, lstm
        if cell_type in layer.name:
            if encoder_cell_index >= len(latent_dim_encoder):
                decoder_cell_pos.append(i)
            else:
                encoder_cell_pos.append(i)
                encoder_cell_index = encoder_cell_index + 1       

    if cell_type == "rnn" or cell_type == "gru":
        encoder_outputs, state = model.layers[encoder_cell_pos[-1]].output
        encoder_model = keras.Model(encoder_inputs, [state])
    
    elif cell_type == "lstm":
        encoder_outputs, state_h_enc, state_c_enc = model.layers[encoder_cell_pos[-1]].output
        encoder_model = keras.Model(encoder_inputs, [state_h_enc, state_c_enc])
    
    else:
        return

    decoder_outputs =  model.layers[decoder_embedding_pos](decoder_inputs)
    for dec in range(len(decoder_cell_pos)):
        
        if cell_type == "rnn" or cell_type == "gru":
            state = keras.Input(shape = (latent_dim_decoder[dec], ))
            current_states_inputs = [state]
            decoder_outputs, state = model.layers[decoder_cell_pos[dec]](decoder_outputs, initial_state = current_states_inputs)
            decoder_states += [state]

        elif cell_type == "lstm":
            state_h_dec, state_c_dec = keras.Input(shape = (latent_dim_decoder[dec],)),  keras.Input(shape = (latent_dim_decoder[dec],))
            current_states_inputs = [state_h_dec, state_c_dec]
            decoder_outputs, state_h_dec,state_c_dec = model.layers[decoder_cell_pos[dec]](decoder_outputs, initial_state = current_states_inputs)
            decoder_states += [state_h_dec, state_c_dec]
        
        decoder_states_inputs = decoder_states_inputs + current_states_inputs

    
    decoder_dense = model.layers[dense_pos]
    decoder_outputs = decoder_dense(decoder_outputs)

    # Building decoder model
    decoder_model = keras.Model(
        [decoder_inputs] + decoder_states_inputs, [decoder_outputs] + decoder_states
    )
    
    for seq_iterator  in range(len(input_texts_val)):     
        #input to encoder   
        '''
        input_seq : input to encoder  
        states_values : hidden states of decoder
        target_seq : expected target
        '''
        input_seq = encoder_test_input_data[seq_iterator  : seq_iterator  + 1]
        states_value = [encoder_model.predict(input_seq)] * len(decoder_cell_pos)
        target_seq = np.zeros((1, 1))
        target_seq[0, 0 ] = target_token_index["\t"]
        flag = False
        decodec_sen = ""

        if beam_size != 0:
        #  inference on the model using Beam Search Decoder

            sequences = [[0.0, 0, states_value, target_seq,  list(),list()]]

            while not flag:

                potential_list = list()
            
                for i in range(len(sequences)):
                    output = decoder_model.predict([sequences[i][3]] + sequences[i][2])
                    output_tokens, states_value = output[0], output[1:]
                    probability = output_tokens[0,-1,:]
              
                    score, word_end, sv, target_temp_seq, seq, d_text = sequences[i]
              
                    if word_end == 0:
                        for j in range(len(reverse_target_char_index)):
                            char = reverse_target_char_index[j]

                            target_seq = np.zeros((1, 1))
                            target_seq[0, 0] = j

                            potential = [score - np.log(probability[j]), 0, states_value, target_seq,  seq + [j] , d_text + [char] ]
                            potential_list.append(potential)
            
                potential_list_order = sorted(potential_list, key=lambda x:x[0])

                min_length = min(beam_size, len(potential_list_order))

                sequences = potential_list_order[:min_length]

                flag = True
           
                for sq in range(len(sequences)):
               
                    score, word_end, sv, target_temp_seq, seq, d_text = sequences[sq]

                    if d_text[-1] == "\n": 
                        word_end = 1

                    if len(d_text) > max_decoder_seq_length : 
                        word_end = 1

                    sequences[sq] = [score, word_end, sv, target_temp_seq, seq, d_text].copy()

                    if word_end == 0: 
                        flag = False

                if sequences[0][-1][-1]=="\n": 
                    flag = True

            decoded_word = ''.join(sequences[0][len(input_texts_val)])

        else:
        # Decode sequence   
            while not flag:
                output = decoder_model.predict([target_seq] + states_value)
                output_tokens, states_value = output[0], output[1:]
                sampled_token_index = np.argmax(output_tokens[0, -1, :])
                sampled_char = reverse_target_char_index[sampled_token_index]
                decoded_word = decoded_word + sampled_char
                if sampled_char == "\n" or len(decodec_sen) > max_decoder_seq_length:
                    flag = True
                target_seq = np.zeros((1, 1))
                target_seq[0, 0] = sampled_token_index 

        
        original_word = target_texts_test[seq_iterator ][1:]
        predictions_vanilla.write(input_texts_test[seq_iterator ] + "," + decoded_word[:-1] + "," + original_word[:-1] + "\n")
        
        # Matched
        if(original_word == decoded_word): 
            count = count + 1

    val_accuracy = count / len(input_texts_test)   
    
    wandb.log( { "val_accuracy": val_accuracy})
    
    
```

```python
sweep_id = wandb.sweep(sweep_config, entity="riya_verma", project="Assignment3_rough_work")
wandb.agent(sweep_id, project = "Assignment3_rough_work", function = train)
```
